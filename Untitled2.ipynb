Masking and Cleansing of audio dataset

```python
def envelope(y , rate, threshold):
    mask=[]
    y=pad.Series(y).apply(nup.abs)
    y_mean= y.rolling(window=int(rate/10) , min_periods=1 , center= True).mean()
    for mean in y_mean:
        if mean>threshold:
            mask.append(True)
        else:
            mask.append(False)
    return mask
```

```python
import pickle
from tqdm import tqdm
from scipy.io import wavfile
for file in tqdm(glob.glob("E:\\5_sem_Mini_Project\\Audio_dataset\\audio_speech_actors_01-24\\Actor_*\\*.wav")):
    file_name = os.path.basename(file)
    signal , rate = librosa.load(file, sr=16000)
    mask=envelope(signal,rate,0.0005)
    wavfile.write(filename="E:\\5_sem_Mini_Project\\filter_speech\\"+str(file_name), rate=rate, data=signal[mask])
    
```

EXTRACTING FEATURE FROM AUDIO DATASET

```python
def extractFeature(fileName, mfcc, chroma, mel):
    with soundfile.SoundFile(fileName) as soundFile:
        X = soundFile.read(dtype="float32")
        sampleRate=soundFile.samplerate
        if chroma:
            stft=nup.abs(librosa.stft(X))
        result=nup.array([])
        if mfcc:
            mfcc=nup.mean(librosa.feature.mfcc(y=X, sr=sampleRate, n_mfcc=40).T, axis=0)
        result=nup.hstack((result,mfcc))
        if chroma:
            chroma=nup.mean(librosa.feature.chroma_stft(S=stft,sr=sampleRate).T, axis=0)
        result=nup.hstack((result,chroma))
        if mel:
            mel=nup.mean(librosa.feature.melspectrogram(X, sr=sampleRate).T,axis=0)
        result=nup.hstack((result,mel))
    return result
```

```python
emotions={
    '01':'neutral',                               
    '02':'calm',
    '03':'happy',
    '04':'sad',
    '05':'angry',
    '06':'fearful',
    '07':'disgust',
    '08':'surprised',
}
observedEmotions={'calm','happy','fearful','disgust'}
```

```python
def loadData(test_size=0.2):
    x,y=[],[]
    for file in glob.glob("E:\\5_sem_Mini_Project\\filter_speech\\*.wav"):
        fileName=os.path.basename(file)
        emotion1=emotions[fileName.split("-")[2]]
        if emotion1 not in observedEmotions:
            continue
        feature=extractFeature(file,mfcc=True,chroma=True,mel=True)
        x.append(feature)
        y.append(emotion1)
    return train_test_split(nup.array(x), y, test_size=test_size, random_state=9)
```

SPLITING OF DATASET FOR TRAINING AND TESTING

```python
xTrain,xTest,yTrain,yTest=loadData(test_size=0.20)
```

```python
print((xTrain.shape[0], xTest.shape[0]))
```

```python
print(f'Features extracted: {xTrain.shape[1]}')
```

USING MLPClassifier

```python
model=MLPClassifier(alpha=0.01, batch_size=256, epsilon=1e-08, hidden_layer_sizes=(300,),learning_rate='adaptive', max_iter=500)
```

FIT/TRAIN THE MODEL

```python
model.fit(xTrain,yTrain)
```

```python
expected_of_y = yTest
yPred=model.predict(xTest)
yPred
```

RESULT ANALYSIS OF MODEL

```python
print(metrics.confusion_matrix(expected_of_y,yPred))
```

```python
print(classification_report(yTest, yPred))
```

```python
accuracy=accuracy_score(y_true=yTest, y_pred=yPred)

print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
audio=pyaudio.PyAudio()

stream = audio.open(format=pyaudio.paInt16, channels=1, rate=44100,input=True, frames_per_buffer=1024)

frames=[]

RATE = 44100
CHUNK = 1024
RECORD_SECONDS= 5
```

```python
p=int(RATE / CHUNK * RECORD_SECONDS)
for i in range(0, p):
    data = stream.read(CHUNK)
    frames.append(data)
    
print("Done recording")

stream.stop_stream()
stream.close()
audio.terminate()

sound_file = wave.open("myrecording.wav","wb")
sound_file.setnchannels(1)
sound_file.setsampwidth(audio.get_sample_size(pyaudio.paInt16))
sound_file.setframerate(44100)
sound_file.writeframes(b''.join(frames))
sound_file.close()
```
